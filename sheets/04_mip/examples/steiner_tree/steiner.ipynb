# Steiner Tree Problem in Unweighted Graphs

```python
import random

import networkx as nx
from solver import Instance, SteinerTreeSolver
```

```python
G = nx.connected_watts_strogatz_graph(30, 4, 0.2)
terminals = random.sample(list(G.nodes()), 5)

positions = nx.layout.kamada_kawai_layout(G)
nx.draw_networkx(
    G,
    pos=positions,
    node_color=["orange" if node in terminals else "grey" for node in G.nodes()],
)
```

```python
solver = SteinerTreeSolver(Instance(G, terminals))
solution = solver.solve()
```

```python
# Draw the solution
assert solution is not None, "No solution found"
nx.draw_networkx_nodes(G, pos=positions, node_color="lightgrey")
nx.draw_networkx_edges(G, pos=positions, style="dashed", edge_color="lightgrey")
nx.draw_networkx(
    solution,
    pos=positions,
    node_color=["orange" if node in terminals else "grey" for node in solution.nodes()],
)
```
